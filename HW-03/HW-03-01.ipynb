# Image RGB Histogram Maker

In the following code, we will be read an image, extract its colors to RGB. Then draw a histogram (the chart of frequency) for each color.

## Import necessary libraies

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

## Read Image
Read an image from a given directory and save it to a variable.

```python
img = mpimg.imread("images/part1.jpg")
plt.imshow(img)
```

## Extract Image RGBs

Extract the saved image's RGB and store them to new variables.

```python
img_R = img.copy()

# in all pixels with (r, g, b), start from index 1 to 3 (exclusive) and set them to zero, like (r, 0, 0)
img_R[:, :, 1:3] = 0
plt.imshow(img_R)
```

```python
img_G = img.copy()

# in all pixels with (r, g, b), start from index 0 to 3 (exclusive) with 2 steps' length
# and set them to zero, like (0, g, 0)
img_G[:, :, 0:3:2] = 0
plt.imshow(img_G)
```

```python
img_B = img.copy()

# in all pixels with (r, g, b), start from index 0 to 2 (exclusive)
# and set them to zero, like (0, 0, b)
img_B[:, :, 0:2] = 0
plt.imshow(img_B)
```

## Extract number of pixels for each color range
For each color image, compute the number of pixels which have a specific color. For example for the red image, if we have three pixels with r value equals to 155, then the corresponding number for 155 will be set to 3.

```python
# ravel method will expand the multidimensional array values to a 1-D array
pixel_colors_R = img_R[:, :, 0:1].ravel()
```

```python
pixel_colors_G = img_G[:, :, 1:2].ravel()
```

```python
pixel_colors_B = img_B[:, :, 2:3].ravel()
```

## Draw histogram for each color image
Draw histogram for each color image using `.hist()` method. That is, draw a chart with color range on X axis and number of pixels with that specific color on Y axis.

```python
hist(pixel_colors_R, bins=256)[2] # getting second return value will only show the histogram
```

```python
hist(pixel_colors_G, bins=256)[2]
```

```python
hist(pixel_colors_B, bins=256)[2]
```

## Conclusion
By checking each computed histogram, we will find out the distribution of the colors. So if the result chart expresses more of that specif color, then the cart will be more inclined to right and vice versa.

So in the given picture, the majority of pixels' color are more redish, which means they have more red color. Also green color is distributed equally between pixels and most of pixels doesn't have much blue color.
